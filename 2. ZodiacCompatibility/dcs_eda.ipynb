```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max.columns', 20)
pd.set_option('display.max.rows', 1000)
pd.set_option("display.float_format", lambda x: "%.2f" % x )
```

### Load the dataset

```python
df = pd.read_csv("dataset.csv")
df
```

### Create a staging dataframe

```python
df1 = df.copy()
df1
```

### Overview the dataframe

```python
df1.shape
```

### Create global variable

```python
zodiac_signs = [
    ['Aries', 'red'], 
    ['Taurus', 'blue'],
    ['Gemini', 'green'], 
    ['Cancer', 'purple'], 
    ['Leo', 'gold'], 
    ['Virgo', 'black'], 
    ['Libra', 'red'], 
    ['Scorpio', 'blue'], 
    ['Sagittarius', 'green'], 
    ['Capricorn', 'purple'], 
    ['Aquarius', 'gold'], 
    ['Pisces', 'black']]
zodiac_signs = pd.DataFrame(zodiac_signs, columns = ['sign', 'color'])
zodiac_signs
```

### Split the Zodiac_combination column to individual colums zx, zy

```python
iZodiac_split = []
for x in df1.index:
    iZodiac_combination = df1.loc[x, "Zodiac_combination"]

    for y in zodiac_signs.index:
        iZodiac_Signs = zodiac_signs.loc[y, "sign"]
        
        if (iZodiac_Signs in iZodiac_combination) :
            df1.loc[x, "Zodiac_combination"] = iZodiac_combination.replace(iZodiac_Signs, ","+iZodiac_Signs+",")

df1["Zodiac_combination"] = df1["Zodiac_combination"].str.strip(",")
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(' ', '')
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(',,,', ',')
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(',,', ',')

df1[['zx', 'zy']] = df1['Zodiac_combination'].str.split(",", expand=True) 
df1
```

### Drop Zodiac_combination and reorder the columns

```python
df1 = df1.drop(columns = ['Zodiac_combination'])
df1 = df1[['zx', 'zy', 'Compatibility_rate']]
df1 = df1.rename(columns = {"Compatibility_rate" : "compatibility"})
df1
```

### Assign new column IsCompatible to indicate whether sign are compatible

```python
df1 = df1.assign(IsCompatible = (df1['compatibility'].apply(lambda x: "True" if x >= 0.5 else "False")))
df1
```

### The Compatibility Chart

```python
plt.figure(figsize = (15,25))
for x in zodiac_signs.index:
    signName = zodiac_signs.iloc[x].values[0]
    signColor = zodiac_signs.iloc[x].values[1]
    df_plot = df1[df1['zx'] == signName].sort_values(by = 'compatibility', ascending = False)
    plt.subplot(6,2,x+1)
    plt.plot(df_plot['zy'], df_plot['compatibility'], color = signColor, marker = 'o', markersize = 5)
    plt.xticks(rotation='vertical')
    plt.title(signName + " Compatibility Chart", fontdict = {'fontsize': 22})
plt.tight_layout()
plt.show()
```

### The compatibility comparison chart

```python
genCompData = []
for x in zodiac_signs.index:
    signName = zodiac_signs.iloc[x].values[0]
    signColor = zodiac_signs.iloc[x].values[1]
    df_plot = df1[df1['zx'] == signName]
    compCount = df_plot[df_plot['IsCompatible'] == 'True']['IsCompatible'].count()
    nonCompCount = df_plot[df_plot['IsCompatible'] == 'False']['IsCompatible'].count()
    compData = []
    compData.append(signName)
    compData.append(compCount/12)
    compData.append(nonCompCount/12)
    genCompData.append(compData)
df_genCompData = pd.DataFrame(genCompData, columns = ["ZodiacSign", "CompatibilityRate", "NonCompatibilityRate"])
df_genCompData = df_genCompData.sort_values(by = 'CompatibilityRate', ascending = False)
df_genCompData = df_genCompData.reset_index(drop = True)
df_genCompData
```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.plot(df_genCompData['ZodiacSign'], df_genCompData['CompatibilityRate'], color = 'green', marker = 'o', markersize = 5)
plt.xticks(rotation='vertical')
plt.grid(True, axis = 'y')
plt.title("Zodiac Compatibility Chart")
plt.subplot(1,2,2)
plt.plot(df_genCompData['ZodiacSign'], df_genCompData['NonCompatibilityRate'], color = 'red', marker = 'o', markersize = 5)
#plt.boxplot(df_genCompData[['CompatibilityRate', 'NonCompatibilityRate']], tick_labels = ['CompatibilityRate', 'NonCompatibilityRate'], notch = True, bootstrap = 1)
plt.xticks(rotation='vertical')
plt.grid(True, axis = 'y')
plt.title("Zodiac Non Compatibility Chart")
plt.tight_layout()
plt.show()
```
