```python
import pandas as pd
```

```python
pd.set_option('display.max.columns', 20)
pd.set_option('display.max.rows', 1000)
pd.set_option("display.float_format", lambda x: "%.2f" % x )
```

### Load The Dataset

```python
df = pd.read_csv("dataset.csv")
df
```

```python
df.shape
```

### Create a staging dataset

```python
df1 = df.copy()
df1
```

### Melt the party colums to one column PARTY and the values VOTES

```python
df2 = df1.melt(id_vars = ['STATE', 'REGION', 'HIGHEST VOTES','TOTAL VOTES COUNTED'], 
               value_vars = ['APC', 'PDP', 'LP', 'NNPP', 'OTHERS'], 
               ignore_index = True,
               var_name='PARTY', value_name = 'VOTES').sort_values(by = 'STATE')
df2
```

### Recreate the party votes % colums as VOTES100

```python
df2 = df2.assign(VOTES100=(df2['VOTES'] / df2['TOTAL VOTES COUNTED']) * 100)
df2
```

### Sort values and reset index

```python
df2 = df2.sort_values(by = ['STATE', 'PARTY'], ascending = [True, True]).reset_index(drop = True)
df2
```

### Rearrange and reorder the columns

```python
df3 = df2[['STATE', 'REGION', 'PARTY', 'VOTES', 'VOTES100' , 'HIGHEST VOTES', 'TOTAL VOTES COUNTED']]
df3
```

### Rename columns to lower, pipe case

```python
df3 = df3.rename(columns = {
    'STATE' : 'state', 
    'REGION' : 'region', 
    'PARTY' : 'party', 
    'VOTES' : 'votes', 
    'VOTES100' : 'votes100', 
    'HIGHEST VOTES' : 'highest_votes', 
    'TOTAL VOTES COUNTED' : 'total_votes_counted'
})
df3
```

```python
df3.shape
```

### Save to dataset_clean.csv

```python
df3.to_csv("dataset_clean.csv", index = False, float_format = "%.2f")
```
