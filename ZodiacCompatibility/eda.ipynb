## Data Cleaning And Standardization

### Create a zodiac dataframe with sign and color

```python
zodiac_signs = [
    ['Aries', 'red'], 
    ['Taurus', 'blue'],
    ['Gemini', 'green'], 
    ['Cancer', 'purple'], 
    ['Leo', 'gold'], 
    ['Virgo', 'black'], 
    ['Libra', 'red'], 
    ['Scorpio', 'blue'], 
    ['Sagittarius', 'green'], 
    ['Capricorn', 'purple'], 
    ['Aquarius', 'gold'], 
    ['Pisces', 'black']]
zodiac_signs = pd.DataFrame(zodiac_signs, columns = ['sign', 'color'])
zodiac_signs
```

### Split the Zodiac_combination column to individual colums zx, zy

```python
iZodiac_split = []
for x in df1.index:
    iZodiac_combination = df1.loc[x, "Zodiac_combination"]

    for y in zodiac_signs.index:
        iZodiac_Signs = zodiac_signs.loc[y, "sign"]
        
        if (iZodiac_Signs in iZodiac_combination) :
            df1.loc[x, "Zodiac_combination"] = iZodiac_combination.replace(iZodiac_Signs, ","+iZodiac_Signs+",")

df1["Zodiac_combination"] = df1["Zodiac_combination"].str.strip(",")
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(' ', '')
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(',,,', ',')
df1['Zodiac_combination'] = df1['Zodiac_combination'].str.replace(',,', ',')

df1[['zx', 'zy']] = df1['Zodiac_combination'].str.split(",", expand=True) 
df1
```

### Drop Zodiac_combination and reorder the columns

```python
df1 = df1.drop(columns = ['Zodiac_combination'])
df1 = df1[['zx', 'zy', 'Compatibility_rate']]
df1 = df1.rename(columns = {"Compatibility_rate" : "compatibility"})
df1
```

### Assign new column IsCompatible to indicate whether sign are compatible

**IsCompatible** will be True if the compatibility value is > 0.5, anything less than is False

```python
df1 = df1.assign(IsCompatible = (df1['compatibility'].apply(lambda x: True if x >= 0.5 else False)))
df1
```

### 1. Zodiac CompatibilityScore

```python
df_totalCompatibiltyMean = pd.DataFrame(df1.groupby(['zx'])['compatibility'].mean()).sort_values(['compatibility'], ascending = False)
df_totalCompatibiltySum = pd.DataFrame(df1.groupby(['zx'])['compatibility'].sum()).sort_values(['compatibility'], ascending = False)
df_totalCompatibiltyCount = pd.DataFrame(df1.groupby(['zx'])['IsCompatible'].sum()).sort_values(['IsCompatible'], ascending = False)


df_totalCompatibilty = df_totalCompatibiltySum.merge(df_totalCompatibiltyCount, on = ['zx'])
#df_totalCompatibilty = df_totalCompatibilty.merge(df_totalCompatibiltyMean, on = ['zx'])
df_totalCompatibilty = df_totalCompatibilty.reset_index()
df_totalCompatibilty = df_totalCompatibilty.rename(columns = {
    "zx" : "zodiacSign",
    "compatibility" : "compatibilitySum",
    #"compatibility_y" : "compatibilityAvg",
    "IsCompatible" : "compatibilityCount"
})

df_totalCompatibilty['compatibilityScore'] = df_totalCompatibilty['compatibilitySum'] * df_totalCompatibilty['compatibilityCount']
df_totalCompatibilty = df_totalCompatibilty.sort_values(['compatibilityScore'], ascending = False)
df_totalCompatibilty
```

#### Visualization

```python
plt_df = df_totalCompatibilty
plt1 = plt.figure(figsize = (5,5))
mea = np.flip(plt_df['zodiacSign'])
dim = np.flip(plt_df['compatibilityScore'])
bar = plt.barh(mea, dim, color = color_palette_50)
plt.bar_label(bar, ["{:.2f}%".format(x) for x in dim], rotation = "horizontal", padding = 5, color = "#000")
plt.title("Compatibility Score")
plt.xlabel("Score %")
plt.ylabel("ZodiacSign")
plt.show()
```

### 2. Zodiac ComparisonChart

```python
plt.figure(figsize = (15,25))
for x in zodiac_signs.index:
    signName = zodiac_signs.iloc[x].values[0]
    signColor = zodiac_signs.iloc[x].values[1]
    plt_df = df1[df1['zx'] == signName].sort_values(by = 'compatibility', ascending = False)
    plt.subplot(6,2,x+1)
    plt.plot(plt_df['zy'], plt_df['compatibility'], color = signColor, marker = 'o', markersize = 5)
    plt.plot([0,12], [0.5, 0.5], color = "#000", alpha = 0.5) # Compatibilty Mark
    plt.xticks(rotation='vertical')
    plt.title(signName + " Compatibility Chart", fontdict = {'fontsize': 22})
plt.tight_layout()
plt.show()
```
